```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
India_df = df[(df['Location'] == 'India') & (df['Time'] >= 1950) & (df['Time'] <= 2024)]
India_df
```

```python
plt.figure(figsize=(10, 6))
plt.hist(India_df['PopTotal'], bins=20, edgecolor='black')
plt.title('Distribution of India\'s Population from 1950 to 2024')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(India_df['Time'], India_df['PopTotal'], color='skyblue')
plt.title('Population of India from 1950 to 2024')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(India_df['Time'], rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

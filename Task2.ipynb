```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic['Sex'] = le.fit_transform(titanic['Sex'])
titanic['Ticket'] = le.fit_transform(titanic['Ticket'])
titanic['Cabin'] = le.fit_transform(titanic['Cabin'])
titanic['Embarked'] = le.fit_transform(titanic['Embarked'])
```

```python
titanic.info()
```

```python
titanic=titanic.fillna(method='bfill',axis=0)
titanic.isnull().sum()
```

```python
titanic
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6)) # Correct the typo 'figzise' to 'figsize'
sns.boxplot(x='Pclass', y='Age', data=titanic)
plt.title('Distribution of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(titanic['Age'], titanic['Fare'], alpha=0.5)
plt.title('Scatter Plot of Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Age'], bins=20, edgecolor='black', alpha=0.7)
plt.title('Histogram of Age in Titanic Dataset')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
